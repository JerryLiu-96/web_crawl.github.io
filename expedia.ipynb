```python
from bs4 import BeautifulSoup
import time, datetime
import pandas as pd
# A bot can automate the collecting data process. A bot will imitate how an user browse a web page, and then acquire those useful information.
# For more information about how to operate a bot, please refer to the documentation of Selenium at https://selenium-python.readthedocs.io/
from kora.selenium import wd as bot
```

```python
url = "https://www.expedia.com/Hotel-Search?adults=2&d1=2022-02-13&d2=2022-02-14&destination=Seattle%20%28and%20vicinity%29%2C%20Washington%2C%20United%20States%20of%20America&endDate=2022-02-14&latLong=47.611245%2C-122.337566&regionId=178307&rooms=1&semdtl=&sort=RECOMMENDED&startDate=2022-02-13&theme=&useRewards=false&userIntent="
bot.get(url)
hotel_urls=[]
results=[]
for i in range(5):
  soup=BeautifulSoup(bot.page_source,'html.parser')
  hotels=soup.find_all('li', class_="uitk-spacing listing uitk-spacing-margin-blockstart-three horizontal")[-20:]
  for hotel in hotels:
    
    hotel_name = hotel.find("h3", class_="uitk-heading-5 is-visually-hidden").text
    
    hotel_area=hotel.find("div", class_="uitk-text overflow-wrap uitk-type-300 uitk-spacing uitk-spacing-padding-blockend-two uitk-text-default-theme").text
    hotel_rating=hotel.find_all("span", class_="pwa-theme--grey-700")[0].text
    num_review = hotel.find_all("span", class_="pwa-theme--grey-700")[1].text
    
    hotel_price=hotel.find("div", class_="uitk-text uitk-type-end uitk-type-200 uitk-text-default-theme").text.replace(" total","")
    
    hotel_score = hotel.find("span", class_="uitk-type-300 uitk-type-bold all-r-padding-one").text.replace("/5","")
    
    hotel_url = hotel.find("a",class_="listing__link uitk-card-link").attrs["href"]
    
    row={'hotel_name':hotel_name,
         'hotel_area':hotel_area,
         'hotel_rating':hotel_rating,
         'hotel_price':hotel_price,
         'hotel_score':hotel_score,
         'num_review':num_review,
         'hotel_url':hotel_url}
    if hotel_url in hotel_urls:
      print("this hotel has already been added.")
    else:
      print(row)
      hotel_urls.append(hotel_url)
      results.append(row)
  time.sleep(5)
  bot.execute_script("window.scrollTo(0, document.body.scrollHeight);")
#bot.close()
df = pd.DataFrame(results)
```
